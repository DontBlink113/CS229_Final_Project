```python
#transformer architecture needs work

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import autocast, GradScaler
import numpy as np
from sklearn.preprocessing import LabelEncoder

device = (
    'cuda' if torch.cuda.is_available()
    else 'mps' if torch.backends.mps.is_available()
    else 'cpu'
)
print(f"using {device}")

#add stroke index to each point, at add mask
def strokes_to_sequence(strokes, max_len=128):
    points = []
    for stroke_id, stroke in enumerate(strokes):
        stroke = np.array(stroke)
        for pt in stroke:
            points.append([pt[0], pt[1], stroke_id])
        if len(points) >= max_len:
            break

    points = points[:max_len]
    seq_len = len(points)

    padded = np.zeros((max_len, 3), dtype=np.float32)
    padded[:seq_len] = points

    # normalize x,y to [0,1] per sample
    if seq_len > 0:
        xy = padded[:seq_len, :2]
        xy_min = xy.min()
        xy_max = xy.max()
        if xy_max > xy_min:
            padded[:seq_len, :2] = (xy - xy_min) / (xy_max - xy_min)

    # normalize stroke_id to [0,1]
    if seq_len > 0:
        max_stroke = padded[:seq_len, 2].max()
        if max_stroke > 0:
            padded[:seq_len, 2] = padded[:seq_len, 2] / max_stroke

    mask = torch.zeros(max_len, dtype=torch.bool)
    mask[:seq_len] = True

    return torch.tensor(padded, dtype=torch.float32), mask


class StrokeDataset(Dataset):
    def __init__(self, chars_dict, label_encoder, max_len=128):
        self.samples = []
        self.max_len = max_len
        for char, samples in chars_dict.items():
            label = label_encoder.transform([char])[0]
            for strokes in samples:
                self.samples.append((strokes, label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        strokes, label = self.samples[idx]
        seq, mask = strokes_to_sequence(strokes, self.max_len)
        return seq, mask, label


class StrokeTransformer(nn.Module):
    def __init__(self, n_classes=500, d_model=64, nhead=4,
                 num_layers=2, max_len=128, dropout=0.1):
        super().__init__()

        self.input_proj  = nn.Linear(3, d_model)
        self.pos_embedding = nn.Embedding(max_len, d_model)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=d_model * 4,
            dropout=dropout,
            batch_first=True
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.classifier  = nn.Linear(d_model, n_classes)

    def forward(self, x, mask):
        B, L, _ = x.shape
        x = self.input_proj(x)
        pos = torch.arange(L, device=x.device).unsqueeze(0)
        x = x + self.pos_embedding(pos)
        pad_mask = ~mask
        x = self.transformer(x, src_key_padding_mask=pad_mask)
        mask_f = mask.unsqueeze(-1).float()
        x = (x * mask_f).sum(dim=1) / mask_f.sum(dim=1)
        return self.classifier(x)


label_encoder = LabelEncoder()
label_encoder.fit(list(train.keys()))

train_dataset = StrokeDataset(train, label_encoder, max_len=128)
val_dataset   = StrokeDataset(val,   label_encoder, max_len=128)

train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True,  num_workers=4, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=512, shuffle=False, num_workers=4, pin_memory=True)

model     = StrokeTransformer(n_classes=500).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
loss_fn   = nn.CrossEntropyLoss()
scaler    = GradScaler()

print(f"train samples : {len(train_dataset):,}")
print(f"val samples   : {len(val_dataset):,}")
print(f"parameters    : {sum(p.numel() for p in model.parameters()):,}")

best_val_acc = 0.0

for epoch in range(30):
    model.train()
    total_loss = 0
    for x, mask, y in train_loader:
        x, mask, y = x.to(device), mask.to(device), y.to(device)
        optimizer.zero_grad()
        with autocast():
            loss = loss_fn(model(x, mask), y)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        total_loss += loss.item()

    model.eval()
    correct = 0
    with torch.no_grad():
        for x, mask, y in val_loader:
            x, mask, y = x.to(device), mask.to(device), y.to(device)
            correct += (model(x, mask).argmax(dim=1) == y).sum().item()

    val_acc = correct / len(val_dataset)
    print(f"Epoch {epoch+1:2d}  loss={total_loss/len(train_loader):.4f}  val_acc={val_acc:.3f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_model.pt')
        print(f"  saved best model ({val_acc:.3f})")

    scheduler.step()

print(f"\nbest val acc: {best_val_acc:.3f}")
```

```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

class DirectionMapCNN(nn.Module):
    def __init__(self, n_classes=500):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(8, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d(4),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, n_classes)
        )

    def forward(self, x):
        x = x.view(-1, 8, 8, 8)  # (batch, 512) → (batch, 8, 8, 8)
        x = self.conv(x)
        return self.classifier(x)

# --- Data (same as logistic regression) ---
# x_train, y_train, x_val, y_val are your 512-dim feature arrays

label_encoder = LabelEncoder()
label_encoder.fit(y_train)

x_tr = torch.tensor(x_train, dtype=torch.float32)
y_tr = torch.tensor(label_encoder.transform(y_train), dtype=torch.long)
x_v  = torch.tensor(x_val,   dtype=torch.float32)
y_v  = torch.tensor(label_encoder.transform(y_val),   dtype=torch.long)

train_loader = DataLoader(TensorDataset(x_tr, y_tr), batch_size=256, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(TensorDataset(x_v,  y_v),  batch_size=512, shuffle=False, num_workers=2, pin_memory=True)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"using {device}")

model     = DirectionMapCNN(n_classes=500).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)
loss_fn   = nn.CrossEntropyLoss(label_smoothing=0.1)
scaler    = torch.amp.GradScaler('cuda')

print(f"parameters: {sum(p.numel() for p in model.parameters()):,}")

best_val_acc = 0.0

for epoch in range(30):
    model.train()
    total_loss = 0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        with torch.amp.autocast('cuda'):
            loss = loss_fn(model(xb), yb)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        total_loss += loss.item()

    model.eval()
    correct = 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            correct += (model(xb).argmax(dim=1) == yb).sum().item()

    val_acc = correct / len(y_v)
    print(f"Epoch {epoch+1:2d}  loss={total_loss/len(train_loader):.4f}  val_acc={val_acc:.3f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_cnn.pt')
        print(f"  saved best model ({val_acc:.3f})")

    scheduler.step()

print(f"\nbest val acc: {best_val_acc:.3f}")
```
